## Unirse al listado oficial

Completa tus datos en la siguiente celda y ejecutala con **Shift+Enter**.
Esto genera un archivo `.md` con tu registro en la carpeta `alumnos/`.


````python
import os
import subprocess

nombre_completo = ' '    # <-- Completar
email = ' '          # <-- Completar
github_user = ' '      # <-- Completar

os.makedirs('alumnos', exist_ok=True)

# Nombre de archivo seguro
clean_name = github_user.lower().replace(' ', '_')
filename = f'alumnos/{clean_name}.md'

content = f"""# Registro Inicial GitHub

- **Alumno**: {nombre_completo}
- **Email**: {email}
- **Github**: @{github_user}

Infraestructura para Ciencia de Datos.
"""

with open(filename, 'w', encoding='utf-8') as f:
    f.write(content)

print(f'✅ Archivo creado localmente: {filename}')
````

## Siguiente paso

Volve a la terminal y ejecuta:

```bash
git add alumnos/

```

> **Nota**: Si es la primera vez que pusheas esta rama, Git te va a pedir que ejecutes:
> 
> `git push --set-upstream origin registro-apellido-nombre`
>
> Esto es normal, solo pasa la primera vez.

Luego abri un **Pull Request** en GitHub hacia `main`.
